Regression Model

```python
df1 = pd.get_dummies(df, columns=['State'])
```

```python
df1
```

```python
X = df1[['R&D Spend','Administration', 'Marketing Spend', 'State_California', 'State_Florida', 'State_New York']]
Y = df1[['Profit']]
```

```python
import statsmodels.api as sm
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
```

```python
model.summary()
```

Solving MultiCollinearity using Cooks Distance and VIF

Cook's Distance

```python
infl = model.get_influence()
summ_df = infl.summary_frame()
```

```python
summ_df.sort_values('cooks_d', ascending=False)
```

```python
infl.plot_influence()
```

From the above ploting index 46,48,49 has highest Cook's distance for best fitting of model drop them from the dataset

**Variance Influencing Factor**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

From the above table Adminstration has the lowest VIF to fit the best model removing Adminstration from the data set

```python
new_X = df1[['R&D Spend', 'Marketing Spend', 'State_California', 'State_Florida', 'State_New York']]
```

```python
new_model = sm.OLS(Y, new_X).fit()
new_predictions = new_model.predict(new_X)
```

```python
new_model.summary()
```

Removing VIF

```python
new_df = df1.drop(df1.index[[46,49,48]])
```

```python
new_X = new_df[['R&D Spend', 'Marketing Spend', 'State_California', 'State_Florida', 'State_New York']]
new_Y = new_df[['Profit']]
```
