Principal Component Analysis


**Dimensionality Reduction with PCA**

Normalizing Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
df_norm = StandardScaler().fit_transform(df1)
```

```python
pca = PCA(n_components=13)
```

```python
principalComponents = pca.fit_transform(df_norm)
```

```python
from matplotlib import pyplot as plt
PC = range(1, pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_, color='blue')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC)
```

```python
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.3, color='blue')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

The most of variance is put in the first 2 components. Since there is not much variance present from 3rd component, lets just the first 2 componets in our analysis. The scatter plot given an indication that there may be 3 clusters present

**Finding the optimal number of clusters**

```python
wcss = []
```

```python
from sklearn.cluster import KMeans
```

```python
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(PCA_components.iloc[:,:3])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

K Clusters
