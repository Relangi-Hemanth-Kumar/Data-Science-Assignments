```python
df_new = df.pivot_table(index='UserId', columns='Title', values='Rating')
df_new
```

Impute those NaNs with 0 values

```python
df_new.fillna(0, inplace=True)
```

```python
df_new.index = df['UserId'].unique()
df_new
```

**Calculating Cosine Similarity between Users**

```python
from sklearn.metrics import pairwise_distances
user1 = 1 - pairwise_distances( df_new.values, metric='cosine')
user1
```

```python
import numpy as np
np.fill_diagonal(user1, 0)
```

#Store the results in a dataframe

```python
user = pd.DataFrame(user1)
user
```

#Set the index and column names to user ids

```python
user.index = df.UserId.unique()
user.columns = df.UserId.unique()
```

**Considering the most similar customer**

```python
def recommend(custID):
    simID = user.loc[custID].idxmax()
    simID_df = df[df['UserId'] == simID].Title
    custID_df = df[df['UserId'] == custID].Title
    return set(simID_df) - set(custID_df)
```

```python
recommend(276813)
```

```python
recommend(507)
```

```python
recommend(3462)
```
