```python
observed = np.array([50, 142, 131, 70, 435, 1523, 1356, 750])
expected = np.array([42.76531299,  146.81287862,  131.11756787, 72.30424052, 442.23468701, 1518.18712138, 1355.88243213, 747.69575948])
```

```python
statistics, p_value = stats.chisquare(observed, expected, ddof = 3)
print("Statistics = ",statistics,"\n",'P_Value = ', p_value)
```

```python
alpha = 0.05
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between the Proportions of Male and Female')
else:
    print('We fail to reject Null hypothesis there is no significance difference between Proportions of Male and Female')
```

We fail to reject Null hypothesis i,e: All Proportions of Male and Female are equal
