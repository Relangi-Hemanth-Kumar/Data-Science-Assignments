```python
#Step3 : Testing of hypothesis
#Null Hypothesis H0 : Means are equals
#Alternative Hypothesis H1: Means are not equal
#significance = 5% = .0.05
#alpha for two tailed test = 0.025
```

```python
from scipy import stats
zcal,pval = stats.ttest_ind(df["Unit A"],df["Unit B"])
print("z calculated value is",zcal.round(3))
print("p calculated value is",pval.round(3))
alpha = 0.025
if pval<=alpha:
    print("h0 is rejected and h1 is accepted i,e: there is a significant difference between means of Unit A & Unit B")
else:
    print("h0 is accepted and h1 is rejected i,e: there is no significant difference between means of Unit A & Unit B")
```

Hence we accept Null Hypothesis, That is there is no significant difference between means of Unit A and Unit B
