**HIERARCHICAL CLUSTERING**

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df1=(df.iloc[:,1:])
df1
```

```python
# Normalized data frame (considering the numerical part of data)
df1_norm = norm_func(df1.iloc[:,1:])
df1_norm
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,12));plt.title('HIERARCHY DENDROGRAM');plt.xlabel('Index');plt.ylabel('Distance')
dendograms=sch.dendrogram(sch.linkage(df1_norm,"complete"))
plt.show()
```

```python
# create clusters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df1_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
df=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
df_norm.head(10)
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters
model=KMeans(n_clusters=5)
model.fit(df_norm)
model.labels_
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object
df['clust']=md
df_norm.head()
```

```python
df.iloc[:,1:7].groupby(df.clust).mean()
```

```python
df.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

**DBSCAN**

```python
df=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
df.head()
```

```python
df.info()
```

```python
df.drop(['ID#','Award?'],axis=1,inplace=True)
df
```

```python
array=df.values
array
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1, min_samples=12)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([df,cl],axis=1)
```
