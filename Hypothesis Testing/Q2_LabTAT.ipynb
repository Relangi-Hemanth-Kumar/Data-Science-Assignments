```python
#Step3 : Testing of hypothesis
#Null Hypothesis H0 : Means are equals
#Alternative Hypothesis H1: Means are not equal
#significance = 5% = .0.05
```

```python
from scipy import stats
t_statistic , p_value = stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
print('p_value =',p_value)
```

```python
alpha = 0.05
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value<=alpha:
    print("h0 is rejected and h1 is accepted i,e: there is  difference in average TAT among the different laboratories at 5% significanceÂ level.")
else:
    print("h0 is accepted and h1 is rejected i,e: there is no  difference in average TAT among the different laboratories at 5% significanceÂ level ")
```
