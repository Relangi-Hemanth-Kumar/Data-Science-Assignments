**HIERARCHICAL CLUSTERING**

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df1=(df.iloc[:,1:])
df1
```

```python
# Normalized data frame (considering the numerical part of data)
df1_norm = norm_func(df1.iloc[:,1:])
df1_norm
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,12));plt.title('HIERARCHY DENDROGRAM');plt.xlabel('x_axis');plt.ylabel('y_axis')
dendograms=sch.dendrogram(sch.linkage(df1_norm,"complete"))
plt.show()
```

```python
# create clusters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df1_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
clusters=pd.DataFrame(hc.fit_predict(df1_norm),columns=['Clusters'])
clusters['Clusters'].value_counts()
```

```python
df1['Clusters']=clusters
```

**K means clustering**

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
```

```python
X = np.random.uniform(0,1,1000)
Y = np.random.uniform(0,1,1000)
df_xy =pd.DataFrame(columns=["X","Y"])
df_xy.X = X
df_xy.Y = Y
df_xy.plot(x="X",y = "Y",kind="scatter")
```

```python
X = np.random.uniform(0,1,1000)
X
```

```python
model1 = KMeans(n_clusters=5).fit(df_xy)

df_xy.plot(x="X",y = "Y",c=model1.labels_,kind="scatter",s=10,cmap=plt.cm.coolwarm)
```

```python
crime=pd.read_csv("crime_data.csv")
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

df_norm = norm_func(crime.iloc[:,1:])


df_norm.head(10)
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters
model=KMeans(n_clusters=5)
model.fit(df_norm)

model.labels_
```

```python
md=pd.Series(model.labels_)
crime['clust']=md
df_norm.head()
```

```python
crime.iloc[:,1:7].groupby(crime.clust).mean()
```

```python
crime.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

**DBSCAN**

```python
crime=pd.read_csv("crime_data.csv")
crime.head()
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime.values
```

```python
array
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.9, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime,cl],axis=1)
```
